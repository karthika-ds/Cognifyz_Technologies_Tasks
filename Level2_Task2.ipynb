```python
# Step1: Import the libraries
import pandas as pd
```

```python
# Load the dataset in dataframe
df = pd.read_csv('/content/Dataset.csv')
df.head(5)
```

```python
df.info()
```

```python
# Step2 : Determine the most common price range among all the restaurants

most_common_price_range = df['Price range'].mode()
print("Most Common Price Range:", most_common_price_range)
```

```python
# Calculate the average rating for each price range.
average_rating = df.groupby('Price range')['Votes'].mean()
print('Average rating by Price range: ',average_rating)
```

```python
# Identify the color that represents the highest average rating among different price ranges.

highest_average_rating = average_rating.idxmax() # function is a method in Pandas that is used to find the index label of the maximum value in a Series
print('Highest average rating: ', highest_average_rating)
```

```python
# Plotting an average rating by price range
import matplotlib.pyplot as plt
average_rating.plot(kind = 'bar')
plt.xlabel('Price range')
plt.ylabel('Aggregate rating')
plt.title('Average rating by price range')
plt.show()
```
