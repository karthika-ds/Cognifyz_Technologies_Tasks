Extract additional features from the existing columns, such as the length of the restaurant name or address.

Create new features like "Has Table Booking"or "Has Online Delivery" by encoding categorical variables.

```python
# Step1 : import the libraries
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('/content/Dataset.csv')
df.head()
```

```python
df.info()
```

```python
# Step 2: Extract additional features from the existing columns, such as the length of the restaurant name or address.
df['Name'] =df['Restaurant Name'].apply(len)
df['Name']
```

```python
# Step 3: Create new features like "Has Table Booking"or "Has Online Delivery" by encoding categorical variables.

df['Has Table booking'] = df['Has Table booking'].map({True : 1, False :0})
df['Has Online Delivery'] = df['Has Online delivery'].map({True: 1, False : 0})
print(df)
```
